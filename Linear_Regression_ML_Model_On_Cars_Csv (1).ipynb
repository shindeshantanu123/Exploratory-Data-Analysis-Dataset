```python
num_data=df.select_dtypes((int,float))
num_data
```

```python
cate_data=df.select_dtypes(object)
cate_data
```

```python
#2
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
for i in cate_data:
    le=LabelEncoder()
    cate_data[i]=le.fit_transform(cate_data[i])
```

```python
cate_data
```

```python
dff=pd.concat([cate_data,num_data],axis=1)
```

```python
dff
```

```python
dff.drop(['symboling','normalized-losses'],axis=1,inplace=True)
```

```python
dff
```

```python
x=dff.iloc[:,:-1]
x
```

```python
y=dff['price']
y
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.30,random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(xtrain,ytrain)
ypred=lr.predict(xtest)
```

```python
from sklearn.metrics import r2_score
r2_score(ytest,ypred)
```

```python
pd.DataFrame(lr.coef_,x.columns)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
print('MAE :',mean_absolute_error(ytest,ypred))
print('MSE :',mean_squared_error(ytest,ypred))
print('RMSE :',np.sqrt(mean_squared_error(ytest,ypred)))
```
